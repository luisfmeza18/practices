## Seleccion de variables:
Una vez segmentados los datos de laptops tipo notebook, con GPU AMD y con sistema operativo Windows, se seleccionan las siguientes variables de interes para analizar aquellas caracteristicas que influyen en el precio del subconjunto de datos de laptops.

**CPU_core**: Tipo de procesador integrado en la laptop.

**Screen_Size_cm**: Tamaño de la pantalla en cm.

**CPU_frequency**: Frecuencia del CPU en GHz.

**RAM_GB**: Capacidad de la memoria del sistema en GB.

**Weight_kg**: Peso de la laptop en kg.

**Price**: Precio de la laptop en dolares.*

```python
notebook_intcpu_win.info()
```

```python
selected_features = ['CPU_core', 'Screen_Size_cm', 'CPU_frequency', 'RAM_GB', 'Weight_kg', 'Price']
laptop_data = notebook_intcpu_win[selected_features]

laptop_data.head(10)
```

```python
laptop_data.info()
```

```python
# Tras identificar un dato faltante, se opta por eliminar el registro completo para esa laptop
laptop_data = laptop_data.dropna()
```

```python
laptop_data.info()
```

## Medidas de tendencia central y dispersion

```python
laptop_data.describe()
```

## Grafica de la distribucion del precio

```python
plt.figure()
sns.histplot(laptop_data['Price'], kde=True)
plt.title("Distribution of Price")
plt.xlabel("Price (dollars)")
plt.ylabel("Frequency")
```

## Grafica de precio contra familia de CPU Core

```python
plt.figure()
sns.boxplot(x='CPU_core', y="Price", data=laptop_data)
plt.title("Price vs CPU Core type")
plt.xlabel("CPU Core")
plt.ylabel("Price")
```

## Matriz de covarianza y de correlacion

```python
cov_matrix = laptop_data.cov()
corr_matrix = laptop_data.corr()
```
